```python
#Day-16 Constructor in python
#Task-Create a class using constructor to store and print student details.

class student:
     def __init__(self,name,age,course):
      self.name=name
      self.age=age
      self.course=course
    
s1=student("Anand",23,"MCA")
print("Name:",s1.name)
print("Age:",s1.age)
print("Course:",s1.course)
```
